```python
import PIL.Image
import PIL.ImageDraw
import face_recognition
```

```python
#Load the jpg file to a numpy array
image = face_recognition.load_image_file("faces.jpg")
```

```python
#find all the faces in the image
face_locations = face_recognition.face_locations(image)
```

```python
number_of_faces = len(face_locations)
print("I found {} face(s) in this picture".format(number_of_faces))
```

```python
#Load the image into a Python Library object so taht we can draw on top of it and display it
pil_image = PIL.Image.fromarray(image)
```

```python
for face_location in face_locations:

  #print the location of each face in rhis image. Each face is a lost of co-ordinates in (top,right,bottom,left) order.
  top,right,bottom,left = face_location
  print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top,left,bottom,right))

  #Let's draw a box around the face
  draw = PIL.ImageDraw.Draw(pil_image)
  draw.rectangle([left,top,right,bottom],outline="red")
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import Image

plt.imshow(pil_image)
```

```python
#Find all facial features in all the faces in the image
face_landmarks_list = face_recognition.face_landmarks(image)
```

```python
#Loop over each face
for face_landmarks in face_landmarks_list:
  #Loop over each facial feature
  for name,list_of_points in face_landmarks.items():

    #print the location of each facial feature in the image
    print("The {} in this face has the following points: {}".format(name,list_of_points))


    #Let's draw a line around the feature
    draw.line(list_of_points,fill="red",width=2)
```
